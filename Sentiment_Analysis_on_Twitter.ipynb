```
# Tokenisation: for chopping character sequence into peices called tokens 
# and throwing away characters such as punctuations
import nltk
nltk.download('punkt')
# from nltk.tokenize import word_tokenize
from nltk.tokenize import TweetTokenizer 
import requests 
from pprint import pprint
# CASE FOLDING
message = dataframe['message'].tolist()

pprint(message[145:150])
```

```
!pip install emoji
import emoji

message_er = []
for tokens in message:
    # print(tokens)
    filtered_sentence = emoji.demojize(tokens)
    message_er.append(filtered_sentence)

pprint(message_er[100:150])
```

```
import re
def remove_punctuation_re(x):
    x = ' '.join(re.sub("https?://\S+","",x).split()) #Removing URLs
    x = ' '.join(re.sub("^@\S+|\s@\S+","",x).split()) #Removing mentions
    x = ' '.join(re.sub(r'[^$\w\s]','',x).split())
    # x = re.sub(r'[0-9]','',x) #Using regular expression to remove numbers
    # x = re.sub(r'[^a-z ]','',x) #Using regular expression to remove special character and accented character
    return x
    
message_p = [remove_punctuation_re(s) for s in message_er]

message_p[100:150]
```

```
# message_tok = [TweetTokenizer.tokenize(sentence) for sentence in message_p]



tweet_tokenizer = TweetTokenizer()
message_tok = []
for sent in message_p:
    message_tok.append(tweet_tokenizer.tokenize(sent))

message_tok[100:150]
```

```
# Removing stopwords
nltk.download('stopwords')
from nltk.corpus import stopwords as sw
stop_words = sw.words("english")

message_sw=[]
for tokens in message_tok:
    filtered_sentence = [w for w in tokens if not w in stop_words]
    message_sw.append(filtered_sentence)

pprint(message_sw[100:150])
```
